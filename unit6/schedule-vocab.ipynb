## 1. Το μορφότυπο Turtle

Θα έχετε παρατηρήσει από τα προηγούμενα εργαστήρια ότι η μορφή N-Triples δεν είναι ιδιαιτέρως φιλική, ειδικά όταν πρέπει να γράψετε τριάδες “στο χέρι”. Για τον λόγο αυτόν θα χρησιμοποιήσετε εδώ τη μορφή **Turtle** που παρέχει συντομέυσεις για μια πιο συμπαγή γραφή των τριάδων.

Λεπτομέρειες για την Turtle μπορείτε να δείτε στο [σύγγραμα του μαθήματος (ενότητα 2.16)](http://hdl.handle.net/11419/1338). Στη συνέχεια δίνονται μερικά συνοπτικά στοιχεία:

* Δεν είναι υποχρεωτικό μια τριάδα να γραφεί στην ίδια γραμμή (μπορεί να καταλαμβάνει περισσότερες).

* Μπορούμε να χρησιμοποιήσουμε συντομογραφικά προθέματα στα URIs με τη βοήθεια του @prefix:

    ```
    @prefix dc: <http://purl.org/dc/elements/1.1/> .
    ```

    ή ακόμα και:

    ```
    PREFIX dc: <http://purl.org/dc/elements/1.1/>
    ```

* Το κατηγόρημα `rdf:type` είναι τόσο κοινό που μπορεί να συντομευθεί με το `a` ως εξής:

    ```
    uni:Student a rdfs:Class .
    ```

* Όταν έχουμε τριάδες με κοινό υποκείμενο μπορούμε να τις συνοψίσουμε με τη βοήθεια του ερωτηματικού ( ; ) π.χ. το:

    ```
    <http://www.bbc.co.uk/programmes/b03jprmh#programme> dc:title "The Firemaker" ;
                                                         po:position 4 .
    ```

    είναι ισοδύναμο με το:

    ```
    <http://www.bbc.co.uk/programmes/b03jprmh#programme> dc:title "The Firemaker" .

    <http://www.bbc.co.uk/programmes/b03jprmh#programme> po:position 4 .
    ```

* Παρομοίως, μπορούμε με το κόμμα ( , ) να συνοψίσουμε τριάδες με κοινό
υποκείμενο και κατηγόρημα, π.χ. το:

    ```
    <http://www.bbc.co.uk/programmes/p00jxl63#programme> po:episode
    <http://www.bbc.co.uk/programmes/b03jprmh#programme>,
    <http://www.bbc.co.uk/programmes/b03jprm9#programme> .
    ```

    είναι ισοδύναμο με το:

    ```
    <http://www.bbc.co.uk/programmes/p00jxl63#programme> po:episode
    <http://www.bbc.co.uk/programmes/b03jprmh#programme> .
    <http://www.bbc.co.uk/programmes/p00jxl63#programme> po:episode
    <http://www.bbc.co.uk/programmes/b03jprm9#programme> .
    ```

## 3. Σύνδεση των δεδομένων με το λεξιλόγιο

Το αρχείο Turtle που φτιάξατε στο προηγούμενο βήμα περιγράφει το σχήμα των δεδομένων σας. Θα πρέπει όμως να **συνδέσετε** τα αρχικά σας δεδομένα με το λεξιλόγιο. Πώς θα γίνει αυτό;

Για κάθε URI των δεδομένων σας, θα πρέπει να προστεθεί (στα δεδομένα πάντα) μια τριάδα της μορφής:

```
<your-uri> rdf:type <type-of-this-uri> .
```

Αυτό πρέπει να γίνει για κάθε URI, δηλαδή

* για τις διαλέξεις (ανώνυμοι κόμβοι στη θέση υποκειμένου)

* για τους διδάσκοντες (URIs σε θέση αντικειμένου)

* για τα μαθήματα (URIs σε θέση αντικειμένου)

* για τις αίθουσες (URIs σε θέση αντικειμένου)
